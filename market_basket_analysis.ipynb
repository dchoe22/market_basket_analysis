# Import libraries 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

# Import CSV

```python
df = pd.read_csv("/Users/dhchoe/Desktop/D212/D212_3/teleco_market_basket.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

# Drop empty rows

```python
df = df.dropna(how='all')
```

```python
df.head()
```

```python
df.shape
```

# Create a list of list 

```python
trans = []
for i in range (0, 7501):
    trans.append([str(df.values[i,j]) for j in range(0, 20)])
```

# Transactionalize dataset

```python
TE = TransactionEncoder()
array = TE.fit(trans).transform(trans)
```

# Convert to dateframe

```python
clean_df = pd.DataFrame(array, columns = TE.columns_)
clean_df
```

```python
for col in clean_df.columns:
    print(col)
```

# Drop empty column(s)

```python
df_clean = clean_df.drop(['nan'], axis=1)
```

```python
for col in df_clean.columns:
    print(col)
```

```python
df_clean.shape
```

# Extract clean csv

```python
df_clean.to_csv('/Users/dhchoe/Desktop/D212/D212_3/clean_df.csv')
```

# Apriori 

```python
a_rules = apriori(df_clean, min_support = 0.05, use_colnames = True)
a_rules.head()
```

# Set metrics and mine rules

```python
ass_r = association_rules(a_rules, metric = 'lift', min_threshold = 1)
ass_r
```

```python
ass_r.shape
```

# Top 3 rules generated by Apriori algo

```python
ass_r_sorted = ass_r.sort_values(by=['lift', 'confidence'], ascending=[False, False])

ass_r_sorted.head(3)
```
